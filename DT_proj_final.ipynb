# 2.

This is to create a heatmap for saved images, in case you are collecting data and then wanting to create heatmap for aggregate of many days, we read from a main folder containing folders of each signals, this can save it to html or image.

```python
def create_heatmap_for_iteration(folder_path, iteration, save_directory):
    traffic_density_list = []
    
    for folder_name in os.listdir(folder_path):
        folder = os.path.join(folder_path, folder_name)
        if os.path.isdir(folder):
            image_files = [f for f in os.listdir(folder) if f.endswith(".png")]
            if iteration <= len(image_files):
                image_path = os.path.join(folder, image_files[iteration - 1])
                image = cv2.imread(image_path)
                dark_red_density, light_red_density, yellow_density, green_density = analyze_traffic_density(image)
                traffic_density_list.append([signal_lat_long[folder_name][0], signal_lat_long[folder_name][1], 
                                             max(dark_red_density, light_red_density, yellow_density, green_density)])
    
    m = folium.Map(location=[12.9072707, 77.58028], zoom_start=14, tiles='CartoDB positron')
    HeatMap(traffic_density_list).add_to(m)
    # m.save(os.path.join(save_directory, f"traffic_density_heatmap_iteration_{iteration}.html"))
    img_data = m._to_png(0.5)
    img = Image.open(io.BytesIO(img_data))
    img.save(f"/Users/anvrao/DTeX/Test_Map/image_{iteration}.png")

for i in range(102):
    create_heatmap_for_iteration("/Users/anvrao/DTeX/DT_results", i, "/Users/anvrao/DTeX/Test_Map")
    
```

```python
import moviepy.video.io.ImageSequenceClip

def images_to_video(image_folder_path, fps, extension, video_name, output_format):
    images = [image_folder_path+'/'+img for img in os.listdir(image_folder_path) if img.endswith(extension)]
    movie_clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(images, fps)
    movie_clip.write_videofile(video_name+output_format)

images_to_video("/Users/anvrao/DTeX/Test_Map", 10, ".png", "test3", ".mp4")
```

```python
import os
import cv2

def resize_images_in_folder(folder_path, target_size=(1366, 768)):
    for filename in os.listdir(folder_path):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            image_path = os.path.join(folder_path, filename)
            image = cv2.imread(image_path)
            if image is not None:
                resized_image = cv2.resize(image, target_size)
                cv2.imwrite(image_path, resized_image)
                print(f"Resized {filename} to {target_size}")

folder_path = "C:/Users/pgnmi/Documents/DT_test_heatmap_hudson_1"
resize_images_in_folder(folder_path)
```

```python
import cv2
import os

def images_to_video(image_folder_path, output_video_path, fps):
    image_files = []
    for i in os.listdir(image_folder_path):
        image_files.append(i)
    first_image = cv2.imread(os.path.join(image_folder_path, image_files[0]))
    height, width, _ = first_image.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))
    for image_file in image_files:
        image_path = os.path.join(image_folder_path, image_file)
        frame = cv2.imread(image_path)
        out.write(frame)
    out.release()
image_folder_path = "C:/Users/pgnmi/Documents/DT_test_heatmap_hudson_1"
output_video_path = "output_video_191.mp4"
fps = 1 
images_to_video(image_folder_path, output_video_path, fps)
```
